```python
x=df.iloc[:,0].values
y=df.iloc[:,1].values
```

```python
x
```

```python
#missing values detection in array,use numpy for array to find null value
np.isnan(x).sum()
```

```python
np.isnan(y).sum()
```

```python
print(len(x))
print(len(y))
```

```python
x = x.reshape(-1, 1)
```

```python
#filling the missing values, sklearn is package has models
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
x=imputer.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.30, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train, y_train)
```

```python
LinearRegression()
```

```python
pred=regr.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(x_train, y_train)
r2 = model.score(x_test, y_test)
```

```python
r2
```

```python

#BIAS
#Overfitting 
#underfitting
#overcome all of this by balanced data to not get over,under, bias in data when the model is trained or test data
#trained data should be more in model usually 80-20(train,test),90-10(train,test)
```

```python
plt.title('Height and Weight correlation i training data')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_train,regr.predict(x_train),color='blue')
plt.show()
```

```python
#prediction of y input
my_hit =[[185]]
my_wt_pred=regr.predict(my_hit)
```

```python
my_wt_pred
```

```python
#calculate R
from sklearn.metrics import mean_squared_error
r2_score=regr.score(x,y)
print(r2_score)
```
