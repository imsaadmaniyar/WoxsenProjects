```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'd:\00 - Woxsen\MBA (Business Analytics)\Term 3 - Subjects\Machine Learning for Business\11Mar25\50_startups.csv')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
# dependent and independent variables
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
# convert categorial to numerical
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
le_x = LabelEncoder()
x[:, 3] = le_x.fit_transform(x[:, 3])
ohe_x = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(), [3])],
    remainder = 'passthrough'
)
x = np.array(ohe_x.fit_transform(x), dtype = int)
```

```python
x
```

```python
# split and train data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 0)
```

```python
# MLR model from sklearn
from sklearn.linear_model import LinearRegression
regress = LinearRegression()
regress.fit(x_train, y_train)
```

```python
# prediction
y_pred = regress.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# model performance - Best fitted line error rate
regress.score(x_train, y_train)
```

```python
regress.score(x_test, y_test)
```

```python
# Lets build the model using backward elimination
# Now we will use linear regression from statemodel library
import statsmodels.api as sm
```

```python
x = np.append(arr = np.ones(shape=(50, 1), dtype = int), values = x, axis = 1) # x = c+x
# x = np.append(arr = x, values = np.ones(shape=(50, 1), dtype = int), axis = 1) # x = x+c
```

```python
x
```

```python
# Backward elimination
x_ov = x[:, [0, 1, 2, 3, 4, 5]] # x_ov = x
regress_ols = sm.OLS(endog = y, exog = x_ov).fit()
regress_ols.summary()
```
